```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.cos(x)
plt.plot(x,y)
plt.savefig("cos.png",dpi=300,format="png")
plt.show()
```
