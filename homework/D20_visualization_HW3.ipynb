```python
import seaborn as sns
import numpy as np
sns.set(style="whitegrid")
```

```python
rs = np.random.RandomState(0)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)

sns.residplot(x, y, color='mediumturquoise', lowess=True)
```

```python
sns.distplot(x, bins=15, kde=False, rug=True)
```

```python
sns.distplot(x, bins=60, kde=False, rug=True)
```
